```python
def make_delta(entry):
    # Handle missing or invalid
    if pd.isna(entry):
        return datetime.timedelta(0)
    
    entry = str(entry).strip()
    
    # Remove the word 'days' if it exists (e.g. "0 days 00:02:15")
    entry = re.sub(r'\d+\s+days\s+', '', entry)
    
    # Only proceed if there’s a colon
    if ':' not in entry:
        return datetime.timedelta(0)
    
    parts = entry.split(':')
    
    if len(parts) == 2:  # MM:SS
        h, m, s = 0, parts[0], parts[1]
    elif len(parts) == 3:  # HH:MM:SS
        h, m, s = parts
    else:
        return datetime.timedelta(0)
    
    try:
        return datetime.timedelta(hours=int(h), minutes=int(m), seconds=int(s))
    except ValueError:
        return datetime.timedelta(0)

for col in ['User Ringing Time', 'User Talk Time', 'User Hold Duration', 'ACW Duration']:
    df_acd[col] = df_acd[col].map(make_delta)
```
